```python
import intake 

# load ssgp catalog
catalog_url = 'http://34.105.233.61:8100/catalog.json'
cat = intake.open_stac_catalog(catalog_url)
```

```python
# look inside
cat
```

```python
# walk through nested catalogs / collections
cat.walk()
```

```python
# pick a sub-catalog
subcat = cat[ 'ssgp-ms' ]
subcat
```

```python
# walk through nested catalogs / collections attached to sub-catalog
subcat.walk()
```

```python
# drop down to another sub-catalog
subcat1 = subcat[ 'spot-ms' ]
subcat1
```

```python
# print attributes / assets
subcat1.items()
```

```python
# grab stac item
item = subcat1[ '3673618101']
item
```

```python
print(type(item._stac_obj))
```

```python
# take a look at asset attached to item
item.image
```

```python
# load asset into dask array
da = item.image.to_dask()
da
```

```python
# chunk dask array
da=da.chunk(dict(band=1, x=2560, y=2560 ) )
da
```

```python
# load red channel in aoi
red=da.isel(band=0,x=slice(2560,5120),y=slice(1280,2560))
red=red.chunk(dict(x=256, y=256 ) )

# convert 16-bit reflectance to floating point
red=red/1000
red
```

```python
# load nir channel in aoi
nir=da.isel(band=3,x=slice(2560,5120),y=slice(1280,2560))
nir=nir.chunk(dict(x=256, y=256 ) )

# convert 16-bit reflectance to floating point
nir=nir/1000
nir
```

```python
# compute ndvi
ndvi =(nir - red ) / ( nir  + red )
ndvi=ndvi.chunk(dict(x=256, y=256 ) ) 
ndvi = ndvi.compute()
```

```python
# show red channel
red.plot.imshow()
```

```python
# show nir channel
nir.plot.imshow()
```

```python
# show ndvi
ndvi.plot.imshow()
```
